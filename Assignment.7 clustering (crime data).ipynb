```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt 
```

```python
df=pd.read_csv("crime_data.csv")
```

```python
df
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)  
```

```python
df_norm = norm_func(df.iloc[:,1:])
df_norm
  
```

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(df.iloc[:,1:]))
data  
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(df_norm, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()     
```

```python
from sklearn.cluster import AgglomerativeClustering 
import warnings 
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=3, linkage='average',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
df['clust']=cluster_labels # creating a  new column and assigning it to new column 
df
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
clf = KMeans(n_clusters=4)
y_kmeans = clf.fit_predict(df_norm)
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
df.info()
```

```python
df1 = df.drop(['Unnamed: 0'],axis=1)  
```

```python
array=df1.values
array 
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array) 
X 
```

```python
dbscan = DBSCAN(eps=2, min_samples=3)
dbscan.fit(X)  
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster']) 
```

```python
cl
pd.set_option("display.max_rows", None) 
```

```python
cl
```

```python
df1 = pd.concat([df,cl],axis=1)  
df1     
```

```python
df1.plot(x="Assault",y ="Murder",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan')    
```

```python
dl = dbscan.labels_ 
```

```python
import sklearn
sklearn.metrics.silhouette_score(X, dl)
```

```python
sklearn.metrics.silhouette_score(X, y_kmeans)
```
